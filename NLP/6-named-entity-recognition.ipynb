```python
import nltk
```

```python
text = "The pyramids of Egypt are among the greatest architectural and cultural achievements of the ancient world. Built thousands of years ago, these monumental structures continue to fascinate historians, archaeologists, engineers, and travelers. They stand as enduring symbols of ancient Egyptian civilization, reflecting its religious beliefs, political power, and advanced knowledge of engineering and mathematics."
```

```python
words = nltk.word_tokenize(text)
words
```

```python
tagged = nltk.pos_tag(words)
tagged
```

```python
nltk.ne_chunk(tagged)
```
