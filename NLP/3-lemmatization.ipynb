## Lemmatization

Lemmatization is a text normalization technique in Natural Language Processing (NLP) that aims to reduce the inflected forms of a word to a common base or dictionary form, known as the lemma

```python
from nltk.stem import WordNetLemmatizer
```

## WordNetLemmatizer
The WordNetLemmatizer is a widely used lemmatization tool provided by the Natural Language Toolkit (NLTK) library in Python.

Core Resource: It is a wrapper around the WordNet lexical database.

WordNet is essentially a large, comprehensive, human-curated electronic dictionary and thesaurus for the English language. It groups words into sets of cognitive synonyms called synsets (sets of words with similar meaning).


This dictionary-based approach gives the WordNetLemmatizer the linguistic knowledge required to determine the correct lemma.

```python
lemmatizer = WordNetLemmatizer() 

lemmatizer.lemmatize("going")     

type(lemmatizer)                    
```
