```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize, wordpunct_tokenize
from nltk.tokenize import TreebankWordTokenizer
```

```python
corpus = "My name is Deepak and I have deep interest in Machine Learning. I am also a student. I'm really passionate about Artificial Intelligence."
```

```python
# Tokenization
## paragraph --> sentence
sent_tokenize(corpus)
```

```python
# sentence --> word
sentence = "Dr. Smith said, 'It's a challenging problem, isn't it? The data shows 99.9% accuracy, but we must re-check the $500,000 budget!'"

word_tokenize(sentence)
```

```python
# consider punctuations too in tokenisation
wordpunct_tokenize(sentence)
```

```python
treebank = TreebankWordTokenizer()

treebank.tokenize(sentence)
```
